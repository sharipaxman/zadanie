Гипотезы:
Источник заявки влияет на конверсию(Некоторые источники приводят более платежеспособных клиентов, чем другие.)
Время до первого звонка влияет на оплату(Чем быстрее оператор позвонит клиенту, тем выше вероятность оплаты)
Время до первого контакта влияет на оплату(чем быстрее оператор дозвонился до клиента, тем выше вероятность оплаты)
Группа операторов влияет на конверсию(некоторые группы операторов работают эффективнее других)
Пол клиента влияет на оплату(мужчины и женщины платят с разной вероятностью)

```python
# Гипотеза 1
top_sources = df["source"].value_counts().head(5).index  # Берем 5 самых популярных источников
source_p = list(combinations(top_sources, 2))  # Создаем пары для попарного сравнения

t_test_source = []
for s1, s2 in source_p:
    group1 = df[df["source"] == s1]["is_paid"]
    group2 = df[df["source"] == s2]["is_paid"]
    p_value = ttest_ind(group1, group2, equal_var=False).pvalue
    t_test_source.append((s1, s2, p_value))

df_source_p_values = pd.DataFrame(t_test_source, columns=["Source 1", "Source 2", "P-value"])
df_source_p_values
```

```python
# Гипотеза 2
t_test_first_call = ttest_ind(
    df[df["is_paid"] == 1]["time_to_first_call"].dropna(),  # Время ожидания у оплативших
    df[df["is_paid"] == 0]["time_to_first_call"].dropna(),  # Время ожидания у неоплативших
    equal_var=False
).pvalue
t_test_first_call
```

```python
# Гипотеза 3
t_test_first_reach = ttest_ind(
    df[df["is_paid"] == 1]["time_to_first_reach"].dropna(),  # Время ожидания у оплативших
    df[df["is_paid"] == 0]["time_to_first_reach"].dropna(),  # Время ожидания у неоплативших
    equal_var=False
).pvalue
t_test_first_reach
```

```python
# Гипотеза 4
top_operators = df["operator_group_name"].value_counts().head(5).index  # ТОП-5 операторов
operator_p = list(combinations(top_operators, 2))  # Создаем попарные сравнения

t_test_operator = []
for o1, o2 in operator_p:
    group1 = df[df["operator_group_name"] == o1]["is_paid"]
    group2 = df[df["operator_group_name"] == o2]["is_paid"]
    p_value = ttest_ind(group1, group2, equal_var=False).pvalue
    t_test_operator.append((o1, o2, p_value))

df_operator_p = pd.DataFrame(t_test_operator, columns=["Operator Group 1", "Operator Group 2", "P-value"])
df_operator_p
```

```python
# Гипотеза 5
t_test_gender = ttest_ind(
    df[df["gender"] == 0]["is_paid"].dropna(),
    df[df["gender"] == 1]["is_paid"].dropna(),
    equal_var=False
).pvalue
t_test_gender
```

1 Гипотеза не подтвердилась(p_value > 0.05).Источник заявки не влияет на конверсию.Разница может быть случайной.
2 Гипотеза подтвердилась(p_value < 0.05).Чем быстрее оператор звонит после заявки, тем выше вероятность оплаты
3 Гипотеза подтвердилась(p_value < 0.05).Чем быстрее оператор дозванивается и устанавливает контакт, тем выше вероятность оплаты.
4 Гипотеза не подтвердилась(p_value>0.05).Группа операторов не влияет на конверсию.Разница может быть случайной.
5 Гипотеза не подтвердилась(p_value > 0.05).Пол клиента не влияет на вероятность оплаты.Разница может быть случайной.
